```python
import pandas as pd
import numpy as np

# assume df is your cleaned DataFrame

agg_df = (
    df.groupby(["household_code", "year"])
    .agg(
        number_of_members=("number_of_members", "mean"),  # usually constant
        avg_age=("birth_year", lambda x: np.mean(2025 - x)),  # average current age
        min_birth_year=("birth_year", "min"),  # oldest member
        max_birth_year=("birth_year", "max"),  # youngest member
        avg_height=("height", "mean"),
        males=("sex", lambda x: (x == 1).sum()),  # 1=male
        # females=("sex", lambda x: (x ==2).sum()),  # 2=female 
        avg_education_level=("education_level", lambda x: x.mode()[0] if not x.mode().empty else np.nan),
        active_members=("activity_status_jan", lambda x: (x == 1).sum()),
        inactive_members=("activity_status_jan", lambda x: (x != 1).sum()),
        marital_status_mode=("marital_status", lambda x: x.mode()[0] if not x.mode().empty else np.nan),
        territory_mode=("territory_code", lambda x: x.mode()[0]),
        residence_mode=("residence_code", lambda x: x.mode()[0])
    )
    .reset_index()
)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
agg_df.year = agg_df.year.astype(str)
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(data=agg_df, x='year', y='avg_age', ci=None, marker='o')
plt.title('📈 Средний возраст членов домохозяйств (2020–2024)')
plt.xlabel('Год')
plt.ylabel('Средний возраст, лет')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=agg_df, x='year', y='males')
plt.title('Распределение количества мужчин в домохозяйствах по годам')
plt.xlabel('Год')
plt.ylabel('Количество мужчин')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(data=agg_df, x='year', y='avg_education_level', marker='o', linewidth=2)
plt.title('🎓 Средний уровень образования домохозяйств (2020–2024)')
plt.xlabel('Год')
plt.ylabel('Средний уровень образования')
plt.show()
```

```python
agg_df["total_members"] = agg_df["active_members"] + agg_df["inactive_members"]
agg_df["share_active"] = agg_df["active_members"] / agg_df["total_members"]

plt.figure(figsize=(8,5))
sns.lineplot(data=agg_df, x='year', y='share_active', marker='o')
plt.title('Доля активных членов в домохозяйствах по годам')
plt.xlabel('Год')
plt.ylabel('Доля активных членов (в %)')

plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=agg_df, x='marital_status_mode', hue='year')
plt.title('💍 Распределение типов семейного статуса по годам')
plt.xlabel('Код семейного статуса (1–разведен, 2–женат, 3–неженат и т.д.)')
plt.ylabel('Количество домохозяйств')
plt.legend(title='Год')
plt.show()
```

```python
region_stats = (
    agg_df.groupby(['year', 'territory_mode'])['number_of_members']
      .mean()
      .reset_index()
)

plt.figure(figsize=(10,6))
sns.lineplot(data=region_stats, x='year', y='number_of_members', hue='territory_mode')
plt.title('📍 Средний размер домохозяйства по регионам (2020–2024)')
plt.xlabel('Год')
plt.ylabel('Среднее количество членов семьи')
plt.legend(title='Регион (territory_mode)')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(agg_df[['number_of_members','avg_age','avg_height',
                'males','avg_education_level','active_members','inactive_members']].corr(),
            annot=True, cmap='viridis', fmt=".2f")
plt.title('🧠 Корреляции между характеристиками домохозяйств')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=agg_df, x='avg_age', y='number_of_members', hue='share_active', palette='coolwarm')
plt.title('🌈 Взаимосвязь возраста, размера семьи и активности')
plt.xlabel('Средний возраст')
plt.ylabel('Размер домохозяйства')
plt.legend(title='Доля активных')
plt.show()
```
