```python
# Преобразуем зависимую переменную в числовой формат (если она еще не числовая)
df1['Удовлетворенность_жизнью_в_целом'] = pd.to_numeric(df1['Удовлетворенность_жизнью_в_целом'], errors='coerce')
```

```python
import statsmodels.formula.api as smf

# Строим модель для удовлетворенности жизнью
model = smf.ols('Удовлетворенность_жизнью_в_целом ~ работает_или_нет + опытрабменее6мес + свмрабиучеб_ухуд_усп	+свмрабиучеб_улучш_усп+Часраб20_30+Часрабдо30+Часрабдо10+рабнесвязсо_спец+раб_частичносвязсоспец+Неработают2+возраст +университет + курсобуч+работает_или_нет * возраст+работает_или_нет * курсобуч', data=df1).fit()

# Выводим результаты модели
print(model.summary())
```

```python
import matplotlib.pyplot as plt

# Преобразуем summary в текст
text = model.summary().as_text()

# Настраиваем фигуру
fig, ax = plt.subplots(figsize=(12, 8))
ax.axis('off')  # убираем рамку и оси

# Добавляем текст на изображение
ax.text(0, 1, text, fontsize=10, family='monospace', va='top')

# Сохраняем в файл
plt.savefig('ols_summary.png', bbox_inches='tight', dpi=300)
plt.close()

print("✅ Фотография сохранена как 'ols_summary.png'")
```

```python
# Преобразуем в категориальную переменную с порядком
df1['Удовлетворенность_жизнью_в_целом'] = pd.Categorical(
    df1['Удовлетворенность_жизнью_в_целом'],
    categories=[1, 2, 3, 4, 5],
    ordered=True
)
```

```python

from scipy import stats

# Создадим два набора данных: работающие и неработающие студенты
working_students = df1[df1['работает_или_нет'] == 1]['Удовлетворенность_жизнью_в_целом']
non_working_students = df1[df1['работает_или_нет'] == 0]['Удовлетворенность_жизнью_в_целом']

# Проведем t-тест для двух независимых выборок
t_stat, p_value = stats.ttest_ind(working_students, non_working_students, nan_policy='omit')

# Выводим результаты
print(f't-статистика: {t_stat}')
print(f'p-значение: {p_value}')

# Интерпретируем результаты
if p_value < 0.05:
    print("Есть статистически значимая разница между работающими и неработающими студентами.")
else:
    print("Нет статистически значимой разницы между работающими и неработающими студентами.")
```
